```python
## relation
def relation(X):
    r = [(x,y) for x in X for y in X if (x-y)%3 == 0]
    return r
X = range(1,9)
r = relation(X)
print(r)
```

```python
# distance from origin
x = np.array([2,4,5,6])
y = np.array([3,9,5,2])
dist = np.linalg.norm(x)
dist1 = np.linalg.norm(y)
print(dist)
print(dist1)
```

```python
## distance from both points
x = np.array([2,4,5,6])
Y = np.array([3,9,5,2])
diff = x-Y
print(np.linalg.norm(diff))
```

```python
x = np.array([2,4,5,6])
print(x+3)
print(3*x)
```

```python
## dot 
x = np.array([2,4,5,6])
y = np.array([3,9,5,2])
print(np.dot(x,y))
```

```python
# cosine similarity
cos_sum = np.dot(x,y)/(np.linalg.norm(x) * np.linalg.norm(y))
print(cos_sum)
```
