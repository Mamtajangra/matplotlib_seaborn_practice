```python
## pie chart
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
my_labels = ["reading","watching","sleeping","dancing"]
plt.pie(y,labels = my_labels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
my_labels = ["reading","watching","sleeping","dancing"]
plt.pie(y,labels = my_labels,startangle = 90)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
my_labels = ["reading","watching","sleeping","dancing"]
my_explode = [0,0.4,0,0]
plt.pie(y,labels = my_labels,explode = my_explode)
plt.show()
```
